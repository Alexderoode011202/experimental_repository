```python
# Gaussian Probability Distribution
import math
from math import e, pi as e, pi
from matplotlib import pyplot as plt
def gauss_distr(x, y ,z) -> float:
    """Calculates the gauss probability distribution.
    :param x: Takes the X variable as a number
    :param y: functions as mean
    :param z: fuctions as standard deviation
    :returns: Float number containing the likelyhood of x
    """

    part_1 = 1/(math.sqrt((2*pi*(z**2))))
    
    part_2 = -(((x-y)**2)/(2* z**2))
    
    return part_1 * e** part_2

print(gauss_distr(50,50,1.5))

# Let's plot a gaussian probability distribution
# Let's keep the median and standard deviation constant
# But let the x iterate over a range from 1 to 100
values_list: list = []
for num in range(1001):
    num = num/10
    values_list.append(gauss_distr(num, 50, 1.5))

new_range = [x/10 for x in range(1001)]
plt.plot(new_range,values_list)
plt.scatter(49.5, gauss_distr(49.5, 50, 1.5))
plt.scatter(48, gauss_distr(48,50,1.5))


plt.axis([44,56,0,0.4])
print(gauss_distr(49.5, 50, 1.5))


   
```
